```python
def Arrays2D2PxVec3(arrays,arrayname,arraytype="PxVec3"): 
    arrays = ["%s(%s)" % (arraytype,",".join([str(a) for a in array])) for array in arrays]
    arrays = "std::vector<%s> %s={%s};" % (arraytype, arrayname,",".join(arrays))
    #print(arrays)
    return arrays; 

def Arrays2D2PxU32(arrays,arrayname,arraytype="PxU32"): 
    arrays = [str(a) for array in arrays for a in array]
    arrays = "std::vector<%s> %s={%s};" % (arraytype, arrayname,",".join(arrays))
    #print(arrays)
    return arrays; 
```

```python
import vtk_tools

def Obj2Dat(fname,scale=1):
    Vtp = vtk_tools.Reader(fname+".obj");

    Coords  = vtk_tools.GetCoords(Vtp)
    Coords  = Coords.astype("float32") 
    nCoords = len(Coords)
    
    Coords -= Coords.min(axis=0)
    Coords /= Coords.max()
    Coords *= scale
    print(Coords.max(axis=0))
    
    Cells = vtk_tools.GetCellsList(Vtp)
    Trias = Cells[0]
    Trias = np.array(Trias).astype("uint32")

    print( "\tnCoords", Coords.shape, "nTrias:", Trias.shape  )

    np.savetxt("%s_verts.dat"%fname, Coords)
    np.savetxt("%s_faces.dat"%fname, Trias)
    
    #print(Coords[:5,:])
    #print(Trias[:,:])

#Obj2Dat("bunny")
#Obj2Dat("teapot")
#Obj2Dat("sphere")
#Obj2Dat("teddy",5.0)
#Obj2Dat("buda2",10.0)
Obj2Dat("stanford_dragon_decimate",10.0)
```
